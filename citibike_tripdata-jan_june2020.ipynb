```python
# Combine Databases Jan - June 2020
citibikejan_june2020 = pd.concat([df_citibike_jan2020, df_citibike_feb2020, df_citibike_march2020, df_citibike_april2020, df_citibike_may2020, df_citibike_june2020])
```

```python
# Ensured data is consistent and clean throughout analysis. 
# (Facts: 137967 rows  15 columns).
citibikejan_june2020.head()
```

```python
# 1. How many trips have been recorded total during the chosen period?
# Confirmed joint Count via individual monthly data above. 
print(citibikejan_june2020.count())
print(citibikejan_june2020.isnull().values.any())
```

```python
# Pay attention to labels via citibikenyc.com/system-data
# Trip Duration (seconds)
# Start Time and Date / Stop Time and Date
# Start Station Name / End Station Name
# Station ID / Station Lat/Long
# Bike ID
# User Type (Customer=24-hour pass or 3-day pass user; Subscriber = Annual Member)
# Gender (Zero=unknown; 1=male, 2=female)
# Year of Birth
```

```python
# Transform Gender to NA/Male/Female from 
citibikejan_june2020['gender'] = citibikejan_june2020['gender'].replace([0,1,2], ["unknown", "male", "female"])
citibikejan_june2020.head()
```

```python
citibikejan_june2020.columns
```

```python
# Export to CSV
citibikejan_june2020.to_csv("citibike_tripdata-jan_june2020.csv")
```

```python
# Query Data. What is the number count of unique bikeid's during this time period? 

bikeid_count = len(citibikejan_june2020["bikeid"].unique())
print(f"Unique bikeid_count : {bikeid_count}")
print("----------")
```

```python
# Query Data. What is the number count of unique start and end stations during this time period? 

startstationid_count = len(citibikejan_june2020["start station id"].unique())
print(f"startstationid_count : {startstationid_count}")
print("----------")

startstationname_count = len(citibikejan_june2020["start station name"].unique())
print(f"startstationname_count : {startstationname_count}")
print("----------")

endstationid_count = len(citibikejan_june2020["end station id"].unique())
print(f"endstationid_count : {endstationid_count}")
print("----------")

endstationname_count = len(citibikejan_june2020["end station name"].unique())
print(f"endstationname_count : {endstationname_count}")
print("----------")
```

```python
# Query Data. What is the gender breakdown of active participants (Male v. Female) for this time period? 

Months = 6

gender_unknown_count = citibikejan_june2020[citibikejan_june2020["gender"] == "unknown"].count()
gender_unknown_average_count = (gender_unknown_count/Months)

print("Total: Unknown Gender User Count")
print(gender_unknown_count["gender"])
print("Monthly Average: Unknown Gender User Count")
print(gender_unknown_average_count["gender"])
print("----------")

gender_male_count = citibikejan_june2020[citibikejan_june2020["gender"] == "male"].count()
gender_male_average_count = (gender_male_count/Months)

print("Total: Male Gender User Count")
print(gender_male_count["gender"])
print("Monthly Average: Male Gender User Count")
print(gender_male_average_count["gender"])
print("----------")

gender_female_count = citibikejan_june2020[citibikejan_june2020["gender"] == "female"].count()
gender_female_average_count = (gender_female_count/Months)

print("Total: Female Gender User Count")
print(gender_female_count["gender"])
print("Monthly Average: Female Gender User Count")
print(gender_female_average_count["gender"])
print("----------")
```
